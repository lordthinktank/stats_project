## VECTORIZE 

```python
from sklearn.feature_extraction.text import CountVectorizer
import re
data_dict = {}
deck_hyph =[]
for x in deck_list:
    deck_hyph.append(' '.join([y.replace(' ','-') for y in x]))
# display(deck_hyph)
vec = CountVectorizer(token_pattern= "\S+")
vec_decks = vec.fit_transform(deck_hyph)
vec_df = pd.DataFrame(vec_decks.toarray(), columns=vec.get_feature_names())
display(vec_df)
vec_df.to_html('vectorized_df.html')
vec_df.to_csv('vectorized_df.csv')
```

# DISTANCE MATRIX

## DBSCAN (Density-based Spectral Clustering of Applications with Noise)and other unsupervised clustering can use non-euclidean, but more specifically, user-defined distance metrics. This allows us to cluster in n-dimensional space using a distance that isn't defined by 'lines' between them. 

## In our case, we will define a distance to be the symmetric difference between two decks, then normalize the matrix for DBSCAN.

### DBSCAN finds very dense clusters of items and uses them to determine the "core" of a cluster. Then, based on the hyper parameters, it allows a margin of other items to be included in the cluster.  You do not choose a number of clusters, it presents it to you. 

```python
def get_dist(one,two):
    common = 0
    for i in range(len(one)):
        if one[i] == 2 and two[i] == 2:
            common += 2
        elif one[i] >= 1 and two[i] >= 1:
            common += 1
    return(60-2*common)
        

raw_mat = vec_df.as_matrix()
# display(raw_matrix)

# dist_mat = np.zeros((len(deck_nums), len(deck_nums)), dtype=float)
# display(dist_mat)

dist_mat = np.array([[get_dist(raw_mat[i],raw_mat[j]) for i in range(len(deck_list))] for j in range(len(deck_list))])
```

```python
#normalize the matrix
print_mat_df = pd.DataFrame(dist_mat/60)
display(print_mat_df)
print_mat_df.to_csv('distance_matrix.csv')
```

# GRID SEARCH.
## This searches through a grid of the hyperparameters and stores all the results. I then look for results with certain cluster numbers, and silhouette score. This is the goodness measure. I've gotten decent clusters with silhouette around 0.2



```python
from sklearn.cluster import DBSCAN
from sklearn import metrics

eps_range = list(np.linspace(0.08,0.4,100))
min_samps_range = [i*2+5 for i in range(20)]

db_list = []
results_list = []

for x in eps_range:
    for y in min_samps_range:
        db_list.append(DBSCAN(algorithm = 'auto', eps=x, min_samples=y, metric = 'precomputed').fit(dist_mat/60))
        

# db = DBSCAN(algorithm = 'auto', eps=0.07, min_samples=7, metric = 'precomputed').fit(dist_mat/60)

for db in db_list:
    core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
    core_samples_mask[db.core_sample_indices_] = True
    labels = db.labels_

    # Number of clusters in labels, ignoring noise if present.
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)

    print('Estimated number of clusters: %d' % n_clusters_)
    print("Silhouette Coefficient: %0.3f"
          % metrics.silhouette_score(dist_mat/60, labels))
    results_list.append([n_clusters_,metrics.silhouette_score(dist_mat/60, labels)])
```
